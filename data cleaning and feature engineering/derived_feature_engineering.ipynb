```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

def validate_and_clean_data(df):
    """
    Comprehensive data validation and cleaning function for restaurant dataset.
    Handles mixed data types, missing values, and ensures data integrity before feature engineering.
    
    Parameters:
    df (pandas.DataFrame): Input dataframe with potential mixed data types
    
    Returns:
    pandas.DataFrame: Cleaned dataframe with proper data types
    dict: Cleaning report showing what was cleaned
    """
    
    cleaning_report = {
        'columns_processed': [],
        'mixed_types_found': [],
        'values_converted': {},
        'missing_values_handled': {},
        'errors_encountered': []
    }
    
    # Create a copy to avoid modifying original data
    df_clean = df.copy()
    
    # Define columns that should be numeric
    numeric_columns = [
        'r_rating', 'r_rating_count', 'avg_price_per_person', 
        'avg_price_per_item', 'f_rating', 'f_rating_count', 'f_price'
    ]
    
    print("=== DATA VALIDATION AND CLEANING REPORT ===")
    print(f"Original dataset shape: {df.shape}")
    print(f"Processing {len(numeric_columns)} numeric columns...")
    
    for col in numeric_columns:
        if col in df_clean.columns:
            print(f"\nProcessing column: {col}")
            cleaning_report['columns_processed'].append(col)
            
            # Check for mixed data types
            original_series = df_clean[col]
            non_numeric_mask = pd.to_numeric(original_series, errors='coerce').isna() & original_series.notna()
            
            if non_numeric_mask.any():
                mixed_values = original_series[non_numeric_mask].unique()
                print(f"  ‚ö†Ô∏è  Mixed types found: {mixed_values}")
                cleaning_report['mixed_types_found'].append({
                    'column': col,
                    'mixed_values': list(mixed_values)
                })
            
            # Convert to numeric with error handling
            try:
                # Step 1: Identify problematic values
                numeric_converted = pd.to_numeric(original_series, errors='coerce')
                
                # Step 2: Count conversions
                before_conversion = original_series.count()
                after_conversion = numeric_converted.count()
                values_lost = before_conversion - after_conversion
                
                if values_lost > 0:
                    print(f"  üîÑ Converted to numeric: {values_lost} non-numeric values became NaN")
                    cleaning_report['values_converted'][col] = values_lost
                
                # Step 3: Handle missing values with appropriate strategy
                missing_count = numeric_converted.isna().sum()
                if missing_count > 0:
                    if col in ['r_rating', 'f_rating']:
                        # For ratings, use median (robust to outliers)
                        fill_value = numeric_converted.median()
                        strategy = f"median ({fill_value:.2f})"
                    elif col in ['r_rating_count', 'f_rating_count']:
                        # For counts, use median
                        fill_value = numeric_converted.median()
                        strategy = f"median ({fill_value:.0f})"
                    else:
                        # For prices, use median
                        fill_value = numeric_converted.median()
                        strategy = f"median ({fill_value:.2f})"
                    
                    numeric_converted = numeric_converted.fillna(fill_value)
                    print(f"  ‚úÖ Filled {missing_count} missing values using {strategy}")
                    cleaning_report['missing_values_handled'][col] = {
                        'count': missing_count,
                        'strategy': strategy
                    }
                
                # Step 4: Apply the cleaned column
                df_clean[col] = numeric_converted
                
            except Exception as e:
                error_msg = f"Error processing {col}: {str(e)}"
                print(f"  ‚ùå {error_msg}")
                cleaning_report['errors_encountered'].append(error_msg)
    
    print(f"\n=== CLEANING SUMMARY ===")
    print(f"Columns processed: {len(cleaning_report['columns_processed'])}")
    print(f"Mixed types found in: {len(cleaning_report['mixed_types_found'])} columns")
    print(f"Values converted: {sum(cleaning_report['values_converted'].values())}")
    print(f"Missing values handled: {sum([v['count'] for v in cleaning_report['missing_values_handled'].values()])}")
    print(f"Errors encountered: {len(cleaning_report['errors_encountered'])}")
    
    return df_clean, cleaning_report
```

```python
def create_robust_restaurant_derived_features(df):
    """
    Create all 18 derived features for restaurant dataset with robust error handling.
    This function includes data validation, cleaning, and comprehensive error handling.
    
    Parameters:
    df (pandas.DataFrame): Input dataframe with restaurant data
    
    Returns:
    pandas.DataFrame: Dataframe with all original columns plus 18 derived features
    dict: Processing report with validation and feature creation details
    """
    
    print("üöÄ Starting robust feature engineering pipeline...")
    
    # Step 1: Validate and clean data
    print("\nüìä Step 1: Data Validation and Cleaning")
    df_processed, cleaning_report = validate_and_clean_data(df)
    
    # Step 2: Create derived features with error handling
    print("\n‚öôÔ∏è Step 2: Creating Derived Features")
    
    feature_report = {
        'features_created': [],
        'features_failed': [],
        'warnings': []
    }
    
    try:
        # BLOCK 1: RESTAURANT PERFORMANCE FEATURES
        print("\n  üè™ Creating Restaurant Performance Features...")
        
        # 1. Revenue Proxy
        try:
            df_processed['RevenueProxy'] = df_processed['r_rating_count'] * df_processed['avg_price_per_person']
            feature_report['features_created'].append('RevenueProxy')
            print("    ‚úÖ RevenueProxy created")
        except Exception as e:
            feature_report['features_failed'].append(f"RevenueProxy: {str(e)}")
            print(f"    ‚ùå RevenueProxy failed: {str(e)}")
        
        # 2. Rating Momentum
        try:
            df_processed['RatingMomentum'] = df_processed['r_rating'] * np.log(df_processed['r_rating_count'] + 1)
            feature_report['features_created'].append('RatingMomentum')
            print("    ‚úÖ RatingMomentum created")
        except Exception as e:
            feature_report['features_failed'].append(f"RatingMomentum: {str(e)}")
            print(f"    ‚ùå RatingMomentum failed: {str(e)}")
        
        # 3. Price Sensitivity
        try:
            # Add safety check for division by zero
            df_processed['PriceSensitivity'] = df_processed['r_rating'] / (df_processed['avg_price_per_person'] + 0.01)
            feature_report['features_created'].append('PriceSensitivity')
            print("    ‚úÖ PriceSensitivity created")
        except Exception as e:
            feature_report['features_failed'].append(f"PriceSensitivity: {str(e)}")
            print(f"    ‚ùå PriceSensitivity failed: {str(e)}")
        
        # 4. Popularity Score
        try:
            df_processed['PopularityScore'] = np.log(df_processed['r_rating_count'] + 1) * df_processed['avg_price_per_person']
            feature_report['features_created'].append('PopularityScore')
            print("    ‚úÖ PopularityScore created")
        except Exception as e:
            feature_report['features_failed'].append(f"PopularityScore: {str(e)}")
            print(f"    ‚ùå PopularityScore failed: {str(e)}")
            
        # BLOCK 2: FOOD ITEM QUALITY FEATURES
        print("\n  üçΩÔ∏è Creating Food Item Quality Features...")
        
        # 5. Item Premium
        try:
            df_processed['ItemPremium'] = df_processed['f_price'] / (df_processed['avg_price_per_item'] + 0.01)
            feature_report['features_created'].append('ItemPremium')
            print("    ‚úÖ ItemPremium created")
        except Exception as e:
            feature_report['features_failed'].append(f"ItemPremium: {str(e)}")
            print(f"    ‚ùå ItemPremium failed: {str(e)}")
        
        # 6. Food Rating Efficiency
        try:
            df_processed['FoodRatingEfficiency'] = df_processed['f_rating'] / (df_processed['f_price'] + 0.01)
            feature_report['features_created'].append('FoodRatingEfficiency')
            print("    ‚úÖ FoodRatingEfficiency created")
        except Exception as e:
            feature_report['features_failed'].append(f"FoodRatingEfficiency: {str(e)}")
            print(f"    ‚ùå FoodRatingEfficiency failed: {str(e)}")
        
        # 7. Item Popularity
        try:
            df_processed['ItemPopularity'] = df_processed['f_rating'] * np.log(df_processed['f_rating_count'] + 1)
            feature_report['features_created'].append('ItemPopularity')
            print("    ‚úÖ ItemPopularity created")
        except Exception as e:
            feature_report['features_failed'].append(f"ItemPopularity: {str(e)}")
            print(f"    ‚ùå ItemPopularity failed: {str(e)}")
        
        # 8. Value Index
        try:
            df_processed['ValueIndex'] = df_processed['f_rating'] * df_processed['f_price']
            feature_report['features_created'].append('ValueIndex')
            print("    ‚úÖ ValueIndex created")
        except Exception as e:
            feature_report['features_failed'].append(f"ValueIndex: {str(e)}")
            print(f"    ‚ùå ValueIndex failed: {str(e)}")
        
        # 9. Is Veg Flag
        try:
            df_processed['IsVeg'] = (df_processed['veg_or_non_veg'] == 'veg').astype(int)
            feature_report['features_created'].append('IsVeg')
            print("    ‚úÖ IsVeg created")
        except Exception as e:
            feature_report['features_failed'].append(f"IsVeg: {str(e)}")
            print(f"    ‚ùå IsVeg failed: {str(e)}")
            
        # BLOCK 3: LABEL INDICATOR FEATURES
        print("\n  üè∑Ô∏è Creating Label Indicator Features...")
        
        # Define label categories
        premium_labels = ['BESTSELLER', "CHEF'S SPECIAL", 'MUST TRY']
        health_labels = ['VEGAN', 'GLUTEN FREE', 'DAIRY FREE', 'FODMAP FRIENDLY', 'Eggless available']
        new_seasonal_labels = ['NEW', 'SEASONAL']
        promo_ineligible_labels = ['Not eligible for coupons', 'Not on Pro']
        
        # 10. Premium Label
        try:
            df_processed['PremiumLabel'] = df_processed['label'].isin(premium_labels).astype(int)
            feature_report['features_created'].append('PremiumLabel')
            print("    ‚úÖ PremiumLabel created")
        except Exception as e:
            feature_report['features_failed'].append(f"PremiumLabel: {str(e)}")
            print(f"    ‚ùå PremiumLabel failed: {str(e)}")
        
        # 11. Health Label
        try:
            df_processed['HealthLabel'] = df_processed['label'].isin(health_labels).astype(int)
            feature_report['features_created'].append('HealthLabel')
            print("    ‚úÖ HealthLabel created")
        except Exception as e:
            feature_report['features_failed'].append(f"HealthLabel: {str(e)}")
            print(f"    ‚ùå HealthLabel failed: {str(e)}")
        
        # 12. New or Seasonal
        try:
            df_processed['NewOrSeasonal'] = df_processed['label'].isin(new_seasonal_labels).astype(int)
            feature_report['features_created'].append('NewOrSeasonal')
            print("    ‚úÖ NewOrSeasonal created")
        except Exception as e:
            feature_report['features_failed'].append(f"NewOrSeasonal: {str(e)}")
            print(f"    ‚ùå NewOrSeasonal failed: {str(e)}")
        
        # 13. Promo Ineligible
        try:
            df_processed['PromoIneligible'] = df_processed['label'].isin(promo_ineligible_labels).astype(int)
            feature_report['features_created'].append('PromoIneligible')
            print("    ‚úÖ PromoIneligible created")
        except Exception as e:
            feature_report['features_failed'].append(f"PromoIneligible: {str(e)}")
            print(f"    ‚ùå PromoIneligible failed: {str(e)}")
            
        print("\n  üìä Creating Aggregation Features...")
        
        # BLOCK 4: AGGREGATION FEATURES
        # 14. Menu Size
        try:
            menu_size = df_processed.groupby('r_id')['f_id'].count().reset_index()
            menu_size.columns = ['r_id', 'MenuSize']
            df_processed = df_processed.merge(menu_size, on='r_id', how='left')
            feature_report['features_created'].append('MenuSize')
            print("    ‚úÖ MenuSize created")
        except Exception as e:
            feature_report['features_failed'].append(f"MenuSize: {str(e)}")
            print(f"    ‚ùå MenuSize failed: {str(e)}")
        
        # 15. Average Item Rating Per Restaurant
        try:
            avg_item_rating = df_processed.groupby('r_id')['f_rating'].mean().reset_index()
            avg_item_rating.columns = ['r_id', 'AvgItemRatingPerRestaurant']
            df_processed = df_processed.merge(avg_item_rating, on='r_id', how='left')
            feature_report['features_created'].append('AvgItemRatingPerRestaurant')
            print("    ‚úÖ AvgItemRatingPerRestaurant created")
        except Exception as e:
            feature_report['features_failed'].append(f"AvgItemRatingPerRestaurant: {str(e)}")
            print(f"    ‚ùå AvgItemRatingPerRestaurant failed: {str(e)}")
        
        # 16. Price Range
        try:
            price_stats = df_processed.groupby('r_id')['f_price'].agg(['min', 'max']).reset_index()
            price_stats['PriceRange'] = price_stats['max'] - price_stats['min']
            price_range = price_stats[['r_id', 'PriceRange']]
            df_processed = df_processed.merge(price_range, on='r_id', how='left')
            feature_report['features_created'].append('PriceRange')
            print("    ‚úÖ PriceRange created")
        except Exception as e:
            feature_report['features_failed'].append(f"PriceRange: {str(e)}")
            print(f"    ‚ùå PriceRange failed: {str(e)}")
            
        print("\n  üîó Creating Interaction Features...")
        
        # BLOCK 5: INTERACTION FEATURES
        # 17. Rating Gap
        try:
            df_processed['RatingGap'] = abs(df_processed['r_rating'] - df_processed['f_rating'])
            feature_report['features_created'].append('RatingGap')
            print("    ‚úÖ RatingGap created")
        except Exception as e:
            feature_report['features_failed'].append(f"RatingGap: {str(e)}")
            print(f"    ‚ùå RatingGap failed: {str(e)}")
        
        # 18. Cuisine-City Count
        try:
            cuisine_city_count = df_processed.groupby(['cuisine', 'city'])['r_id'].nunique().reset_index()
            cuisine_city_count.columns = ['cuisine', 'city', 'CuisineCityCount']
            df_processed = df_processed.merge(cuisine_city_count, on=['cuisine', 'city'], how='left')
            feature_report['features_created'].append('CuisineCityCount')
            print("    ‚úÖ CuisineCityCount created")
        except Exception as e:
            feature_report['features_failed'].append(f"CuisineCityCount: {str(e)}")
            print(f"    ‚ùå CuisineCityCount failed: {str(e)}")
            
    except Exception as e:
        print(f"‚ùå Critical error in feature engineering: {str(e)}")
        
    # Final report
    print(f"\n‚úÖ FEATURE ENGINEERING COMPLETE!")
    print(f"Features successfully created: {len(feature_report['features_created'])}")
    print(f"Features that failed: {len(feature_report['features_failed'])}")
    
    # Combine reports
    final_report = {
        'data_cleaning': cleaning_report,
        'feature_engineering': feature_report,
        'final_shape': df_processed.shape,
        'total_features': len(df_processed.columns)
    }
    
    return df_processed, final_report
```

```python
# Step 1: Load your data
df = pd.read_csv('restaurant_side_final.csv')  # Replace with your file path

# Step 2: Check for data issues BEFORE feature engineering
print("Before processing:")
print(df.dtypes)
print(df.info())

# Step 3: Run the robust feature engineering pipeline
df_with_features, processing_report = create_robust_restaurant_derived_features(df)

# Step 4: Verify results
print("\nAfter processing:")
print(f"Original columns: {df.shape[1]}")
print(f"Final columns: {df_with_features.shape[1]}")
print(f"Features added: {df_with_features.shape[1] - df.shape[1]}")

# Step 5: Save cleaned data
df_with_features.to_csv('restaurant_data_with_features.csv', index=False)
```
