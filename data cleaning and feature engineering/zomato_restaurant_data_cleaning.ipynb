```python
import pandas as pd

food_df = pd.read_csv('food.csv')
print("First 5 rows of food.csv:")
display(food_df.head())

menu_df = pd.read_csv('menu.csv')
print("\nFirst 5 rows of menu.csv:")
display(menu_df.head())

restaurant_df = pd.read_csv('restaurant.csv')
print("\nFirst 5 rows of restaurant.csv:")
display(restaurant_df.head())

zomato_dataset_df = pd.read_csv('zomato_dataset.csv')
print("\nFirst 5 rows of zomato_dataset.csv:")
display(zomato_dataset_df.head())
```

```python
df = restaurant_df.copy()
display(df.head())
```

```python
columns_to_drop = ['Unnamed: 0', 'lic_no', 'link', 'address', 'menu']

df = df.drop(columns=columns_to_drop)

print("\nFirst 5 rows of df after dropping columns:")
display(df.head())
```

```python
new_column_names = ['r_id', 'r_name', 'city', 'r_rating', 'r_rating_count', 'avg_price_per_person', 'cuisine']

df.columns = new_column_names

print("\nFirst 5 rows of df after renaming columns:")
display(df.head())
```

```python
menu_subset = menu_df[['r_id', 'f_id', 'price']].rename(columns={'price': 'avg_price_per_item'})

# Merge df with menu_subset on 'r_id'
df = pd.merge(df, menu_subset, on='r_id', how='left')

print("\nFirst 5 rows of df after merging with menu_df:")
display(df.head())
```

```python
food_subset = food_df[['f_id', 'item', 'veg_or_non_veg']].rename(columns={'item': 'f_name'})

# Merge df with food_subset on 'f_id'
df = pd.merge(df, food_subset, on='f_id', how='left')

print("\nFirst 5 rows of df after merging with food_df:")
display(df.head())
```

```python
zomato_dataset_df = zomato_dataset_df.rename(columns={'Restaurant_Name': 'r_name',
                                                        'Dining_Rating': 'f_rating',
                                                        'Dining Votes': 'f_rating_count',
                                                        'Place_Name': 'locality',
                                                        'Item_Name': 'f_name',
                                                        'Best_Seller': 'label',
                                                        'Prices': 'f_price'})


print("\nFirst 5 rows of zomato_dataset_df after renaming columns:")
display(zomato_dataset_df.head())
```

```python
zomato_dataset_df = zomato_dataset_df.drop(columns=['Delivery_Rating', 'Delivery_Votes'])

print("\nFirst 5 rows of zomato_dataset_df after dropping Delivery_Rating and Delivery_Votes:")
display(zomato_dataset_df.head())
```

```python
zomato_dataset_df = zomato_dataset_df.rename(columns={'Cuisine': 'cuisine',
                                                        'City': 'city',
                                                     'Cuisine': 'cuisine'})


print("\nFirst 5 rows of zomato_dataset_df after renaming columns:")
display(zomato_dataset_df.head())
```

```python
# Select relevant columns from zomato_dataset_df for merging
zomato_merge_cols = zomato_dataset_df[['r_name', 'city', 'f_name', 'f_rating', 'f_rating_count', 'locality', 'label', 'f_price']]

# Merge df and zomato_merge_cols
merged_df = pd.merge(df, zomato_merge_cols, on=['r_name', 'city', 'f_name'], how='left', indicator=True)

# Identify rows that are only in the zomato_dataset_df
zomato_only_rows = merged_df[merged_df['_merge'] == 'right_only']

# Drop the merge indicator column
merged_df = merged_df.drop(columns=['_merge'])

# Merge the original df with zomato_only_rows (excluding columns already present in df)
zomato_only_rows = zomato_only_rows.drop(columns=['_merge'])

# Identify columns in zomato_only_rows that are not in df
cols_to_add = [col for col in zomato_only_rows.columns if col not in df.columns]

# Create a subset of zomato_only_rows with only the columns to add and the merge keys
zomato_only_subset = zomato_only_rows[['r_name', 'city', 'f_name'] + cols_to_add]

# Concatenate the original df with the unique rows from zomato_only_rows
df = pd.concat([df, zomato_only_subset], ignore_index=True)

print("\nFirst 5 rows of the merged and concatenated df:")
display(df.head())

print("\nInformation about the final df:")
df.info()
```

```python
df = df.dropna(subset=['f_name'])

print("\nFirst 5 rows of df after removing rows with NaN 'f_name':")
display(df.head())

print("\nInformation about the final df after removing rows with NaN 'f_name':")
df.info()
```

```python
df = df.drop_duplicates()

print("\nFirst 5 rows of df after merging exact duplicates:")
display(df.head())

print("\nInformation about the final df after merging exact duplicates:")
df.info()
```

```python
import numpy as np
# Create a series with the first non-null locality per city
first_locality_per_city = df.groupby('city')['locality'].transform(lambda x: x.dropna().iloc[0] if not x.dropna().empty else np.nan)

# Fill NaN values in locality with the first non-null locality per city
df['locality'] = df['locality'].fillna(first_locality_per_city)
```

```python
import numpy as np

def map_ratings_to_random_number(value):
    if value == 'Too Few Ratings':
        return np.random.randint(0, 21)      # 0 to 20 inclusive
    elif value == '20+ ratings':
        return np.random.randint(20, 51)     # 20 to 50 inclusive
    elif value == '50+ ratings':
        return np.random.randint(50, 101)    # 50 to 100 inclusive
    elif value == '100+ ratings':
        return np.random.randint(100, 501)   # 100 to 500 inclusive
    elif value == '500+ ratings':
        return np.random.randint(500, 1001)  # 500 to 1000 inclusive
    elif value == '1K+ ratings':
        return np.random.randint(1000, 5001) # 1000 to 5000 inclusive
    elif value == '5K+ ratings':
        return np.random.randint(5000, 10001) # 5000 to 10000 inclusive
    else:
        return np.nan

# Apply the function to the 'r_rating_count' column
df['r_rating_count'] = df['r_rating_count'].apply(map_ratings_to_random_number)
```

```python
import numpy as np

# Step 1: Replace '--' with NaN
df['r_rating'] = df['r_rating'].replace('--', np.nan)

# Step 2: Convert to float
df['r_rating'] = df['r_rating'].astype(float)

# Step 3: Fill NaN with random float between 1 and 5, rounded to one decimal
nan_mask = df['r_rating'].isna()
random_ratings = np.round(np.random.uniform(1, 5, size=nan_mask.sum()), 1)
df.loc[nan_mask, 'r_rating'] = random_ratings
```
